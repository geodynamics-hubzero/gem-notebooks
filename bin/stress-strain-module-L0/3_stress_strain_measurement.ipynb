## Stresses within the lithosphere

The stresses in the lithosphere are governed by a combination of *internal* forces and its rheological properties.
Some of the key forces affecting the lithospheric stresses are:

1. **Plate tectonic forces:** large-scale forces as the tectonic plates move relative to each other. These forces generate compressive stress at convergent boundaries, tensional stress at divergent boundaries and shear stress at transform boundaries.
2. **Local dynamics:** small-scale stresses arising due to lateral variations in buoyancy.
3. **Isostatic forces:** e.g., local stress changes due to vertical isostatic adjustments of the lithosphere.
4. **Thermal stresses:** contraction and expansion of rocks due to temperature variations in the lithosphere, e.g., due to radiogenic heat generation, cooling of oceanic lithosphere, etc.



An estimation of the stress state in the lithosphere is therefore important for a better understanding of the plate driving forces, and the forces governing earthquake generation. The stress field is commonly represented using the maximum horizontal compressive stress direction $S_{Hmax}$. The orientation of $S_{Hmax}$ is usually estimated using borehole measurements and focal mechanism inversions. A recent compilation map showing $S_{Hmax}$ directions is shown below.

## Strain response to stresses

The lithosphere responds and deforms under a given stress state. The factors controlling its response include:

1. **Rock strength**: the rock properties at a given temperature and pressure conditions can resist or promote deformation.
2. **Brittle-ductile transition depth**: the brittle upper crust fails by fracturing (earthquakes), whereas the ductile lower crust/upper mantle flows under stress.
3. **Fluids**: presence of fluids reduces the strength of the rocks promoting fracturing and slip.
4. **Local heterogeneities**: weak zones, such as pre-existing faults, can localize deformation.



Strains on Earth can usually be calculated using geodetic instruments that record how a given point on the Earth's surface moves over time. We use strain rates instead of total strain, since how fast or slow a given regions is deforming is given by the strain rate.

## Earthquakes
Earthquakes are a direct result of elastic stress accumulation and its sudden release as the stress exceeds the rock's strength. This accumulation of stress and strain is also referred to as *elastic rebound theory*.
Due to higher tectonic movements leading to stress buildup, earthquakes primarily occur along the plate boundaries (or high-strain regions) as shown in the figure below.
