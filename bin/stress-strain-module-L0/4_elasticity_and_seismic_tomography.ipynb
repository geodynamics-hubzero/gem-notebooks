## Elasticity and solid-state geophysics

The elastic behavior of an isotropic, ideally elastic, solid at infinite frequency is described by two independent elastic moduli, called the Lame's parameters $K$ and $\mu$. These elastic constants relate the stress-strain relationship in the solid, see more in the notebook [Stress-strain relationship](./2_stress_strain_relationships.ipynb).

The elastic moduli are related to the seismic wave velocities in the solid, the compressional wave velocity $V_p$ and the shear wave velocity $V_s$, as follows:

$$ V_p^2 = \dfrac{K + 4\mu/3}{\rho} \quad, \quad V_s^2 = \dfrac{\mu}{\rho}$$

where $\rho$ is the density of the solid.

### Elastic moduli and relation to physical state

From mineral physics experiments, we have $K (P, T, X)$ and $\mu (P, T, X)$, at a given pressure $P$, temperature $T$, and mineral composition $X$. Since we also have seismic wave velocities from the earthquake data, we can estimate which composition/temperature combination matches the observed wave velocities at a given depth (pressure).

**Sensitivity to temperature**
- Increasing temperature decreases both $K$ and $\mu$, and thus decreases both $V_p$ and $V_s$.
- ${d\mu}/{dT}$ is larger than ${d K}/{dT}$, suggesting that S-wave velocities are stronger indicators for temperature variations than P-wave velocities.
- Example: low $V_s$ in asthenosphere, plumes are interpreted as hotter regions in the mantle, and high $V_s$ in subducting slabs are interpreted as colder regions.

**Sensitivity to composition**

Mineral composition effects elastic properties and hence seismic velocities. For example, continental crust has lower seismic velocities than oceanic crust, because of its felsic composition (more silica).
