## Viscosity under Composite Creep

Experimental studies on rock deformation such as those done by [Hirth and Kohlstedt (2003)](https://doi.org/10.1029/138GM06) and [Karato and Wu (1993)](https://doi.org/10.1126/science.260.5109.771) have shown that mantle-rheology is **composite**. This means that the total strain rate is the sum of the strain rates from different deformation mechanisms, primarily *diffusion creep* and *dislocation creep*.

The mantle viscosity for composite creep can be expressed as:

$$\eta^\text {i} = A d^m \sigma^{1-n} \exp \left( \dfrac{E_a^i + PV_a^i}{n^iRT} \right) \tag{1}$$

where, $R$ is the gas constant, $m$ is the grain-size exponent, and $n$ is the stress exponent. $E$, $V$, $A$ are respectively the activation energy, the activation volume and a pre-exponential factor (accounting for all other effects on mantle rheology, such as water and melt content) for the respective rheological mechanism ($i$) considered (diffusion or dislocation creep).

### Effective viscosity

From eq.(1), we can see that the viscosity is highly nonlinear depending on stress, temperature, pressure, and grain size, assuming a given water content. The effective viscosity of the mantle can then be given as

$$\eta_{eff} = \dfrac{\sigma}{2 (\dot{\varepsilon}^\text{diff} + \dot{\varepsilon}^\text{disl})}$$

This suggests that the fastest deformation mechanism will dominate the mantle deformation under given physical conditions.

### Dominant mechanism

We can compute the dominant mechanism by looking at transition stress, $\sigma_t$, defined as when the strain-rates due to the two different mechanisms are equal (i.e., $\dot{\varepsilon}^\text{diff} = \dot{\varepsilon}^\text{disl}$), and above which dislocation creep dominates. By equating the strain rates from both mechanisms, we can derive an expression for the transition stress:

$$\sigma_t = \left(\dfrac{A_{diff}}{A_{disl}}\right)^{\frac{1}{n-1}} d^{\frac{m}{n-1}} \exp \left( \dfrac{(E_{disl} - E_{diff}) + P(V_{disl} - V_{diff})} {RT} \right) ^\frac{1}{n-1}, \tag{2} $$

The laboratory-derived experiments have shown that $E_{diff} < E_{disl}$ and $V_{diff} < V_{disl}$ (Hirth and Kohlstedt, 2003), which means that $\sigma_t$ decreases with increasing temperature from eq. 2. This implies that diffusion creep is more dominant in the hotter and deeper parts of the mantle, whereas dislocation creep is more dominant in the uppermost-mantle and regions of high stress (e.g., lithosphere).

### Earth's Observations and Rheology

**Seismic Anisotropy**

Dislocation creep, dominant in the upper mantle, leads to lattice-preferred orientation (LPO) of olivine minerals. This LPO is observed through seismic anisotropy, where seismic waves travel at different speeds depending on their direction through the medium, and is a key indicator of active mantle flow.

**Geoid**

The observed geoid anomalies, which are sensitive to the mantle buoyancy and viscosity structure, have been studied using global mantle convection models. Most studies are able to fit the long-wavelength geoid using a simple radial- and temperature-dependent viscosity structure, suggesting that dislocation creep can be ignored for these large scales.
