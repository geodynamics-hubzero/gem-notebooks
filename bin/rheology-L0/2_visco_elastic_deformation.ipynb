## Linear Viscoelastic Model

The Earth deforms both elastically and viscously at different time scales:
* Elastic deformation occurs instantaneously.
* Viscous deformation increases with time.

A viscoelastic model captures both behaviors. Linear viscoelasticity assumes that strains generated in the model are linearly proportional to the applied stresses. Such a model is widely used to model viscoelastic creep due to lithospheric bending, see [an example here](../elasticity-flexure-L0/2_bending_of_lithosphere_under_line_loads.ipynb).

### Fundamental Characteristics
All visco-elastic materials show three fundamental behaviors as they are deformed:
* Creep: Increase in strain over time as a constant stress is applied on the material, i.e., $\frac{d\sigma}{dt} = 0$
* Relaxation: Decrease in stress over time as a constant strain is applied on the material, i.e., $\frac{d\epsilon}{dt} = 0$
* Recovery: Decrease in strain over time after the stress is removed, i.e., $\sigma = 0$


### Mechanical Models
Mechanical models are used to represent the behavior of viscoelastic materials. Some common models include:

1. **Maxwell Model**: Consists of a spring (elastic element) and a dashpot (viscous element) in series. This leads to a constant stress across the both components and the total strain of the model is equal to it's elastic and viscous constituents. It is widely used to model viscoelastic relaxation.

   Constitutive equation:
   \begin{align}
   \dot{\varepsilon_M} = \dfrac{\sigma_M}{\eta} + \dfrac{\dot{\sigma_M}}{E} \end{align}
   
   where $E$ and $\eta$ is the elastic moduli and viscosity, respectively, and $\dot{\varepsilon_M}$ and $\dot{\sigma_M}$, are the time derivatives of Maxwell strain and stress, respectively.

2. **Kelvin-Voigt Model**: Consists of a spring and a dashpot in parallel. This leads to a constant strain across both components and the total stress of the model is equal to its elastic and viscous constituents.
    
    Constitutive equation:
    \begin{align}
    \sigma_{KV} = E \varepsilon_{KV} + \eta \dot{\varepsilon_{KV}}
    \end{align}
    
     where $E$ and $\eta$ is the elastic moduli and viscosity, respectively, and $\varepsilon_{KV}$ and $\dot{\varepsilon_{KV}}$ are theKkelvin-Voigt strain and its time derivative, respectively.

3. **Standard Linear Solid**: Consists of a spring and a Maxwell arm in a parallel configuration. The total stress of the system is the sum of stresses across the Maxwell and the spring arms, whereas the strain is equal in both arms.

   Constitutive equation:
   $${\dot\sigma_{LS}} + \dfrac{E_2}{\eta} \sigma_{LS}  = (E_1 + E_2)\dot{\varepsilon_{LS}} + \dfrac{E_1 E_2}{\eta} \varepsilon_{LS}$$
   where $E_1$, $E_2$ are the elastic moduli in the Maxwell and spring element, respectively, and $\sigma_{LS}$ and $\varepsilon$ are the stress and strain in the standard linear solid model, respectively, and $\dot{\sigma_{LS}}$ and $\dot{\varepsilon_{LS}}$ are their time derivatives.

### Transient Viscoelastic Response

For Earth's viscoelastic behavior, two limits exist that describes the observed Glacial Isostatic Adjustment response: the unrelaxed, zero dissipative, elastic limit that occurs instantaneously and the relaxed, fully dissipative, viscous limit that occurs after an infinite time.

Maxwell model captures both limits and the transition in between through the characteristic relaxation time, $\tau = \frac{\eta}{E}$, such that elastic limit dominates at times <<$\tau$ and viscous flow at time >>$\tau$. However, the Maxwell model does not capture the transient creep before the steady-state viscous flow.

To capture transient components, one or more Kelvin-Voigt component(s) may be added, in total known as a **Burgers model**, resulting in additional relaxation time on top of Maxwell model. The following figure shows the transient response of the Burger's and the Maxwell model.
