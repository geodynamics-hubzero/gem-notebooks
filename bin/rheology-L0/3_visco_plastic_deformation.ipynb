## Plastic rheology

Plastic deformation is irreversible and permanent, and occurs when the stress exceeds a certain yield stress. Under the yield stress, the materials behave either viscously or elastically, depending on the timescale of the applied stress.

### Viscoplasticity

We can assume the total strain rate, $\dot{\varepsilon}$, into purely viscous and plastic parts:

$$\dot{\varepsilon} = \dot{\varepsilon}_{v} + \dot{\varepsilon}_{p} \tag{1} $$

where, $\dot{\varepsilon_p}$ is defined in terms of a plastic multiplier $\dot{\lambda}$ and a scalar plastic potential function $g$:

$$\dot{\varepsilon}_{p} = \dot{\lambda} \dfrac{\partial g}{\partial \tau} \tag{2}$$

To derive a plastic viscosity, we assume that deviatoric stresses arise purely from the viscous component of the strain rate, that is

$$\tau = 2 \eta \dot{\varepsilon_v} \tag{3} $$

where, $\tau$ is the deviatoric stress and is bounded by the yield stress, $Y$, before failure ($\tau <= Y$), and $\dot{\varepsilon_v}$ is the viscous component of the second invariant of the strain rate tensor.

Noting, ${\partial g}/{\partial \tau} = \frac{1}{2} \tau_{II}^{-1} \tau$ and substituting equations (1) and (2) into equation (3), we get

$$ \tau = 2 \eta \left(\dot{\varepsilon} - \dot{\lambda} \dfrac{\tau}{2\tau_{II}} \right)$$



### Yielding

The yield surface, $f(\mathbf{\sigma})$ is a scalar function defining the failure (yield) state of a material. Yield surfaces are assumed to be of the following form:

$f(\mathbf{\sigma}) = \tau_{II} - Y(\mathbf{\sigma})$,

where the scalar function, $Y$, defines the yield criterion. For materials not yielding, $f < 0$, and at yield, $f = 0$ or $\tau_{II} = Y$.

Following are the commonly used yield criteria:

1. von Mises Plasticity: $Y = C$, where $C$ is a constant cohesion independent of pressure, and is suitable for modeling ductile failure.

2. Drucker-Prager Plasticity: $Y = P \sin \phi + C \cos \phi$. Here, $\phi$ is the angle of internal friction, $C$ is the cohesion, and depends on the confining pressure $P$.

3. Mohr-Coulomb Plasticity: $Y = C + \tan \phi \sigma_n$, where $\sigma_n$ is the normal stress on the failure plane and other parameters are as defined above. This criterion is suitable to represent brittle failure.

### Types of plastic failure

At yielding, rocks fail can either in a brittle or ductile manner.

* At low confining pressures, *brittle* failure occurs as the rock fractures and its strength decreases.
* At high confining pressures, *ductile* flow occurs as plastic strain accumulates without any significant change to the rock's strength.

### Brittle-Ductile Transition (BDT)
The depth defines the state, i.e., temperature and pressure conditions, at which rocks transition from brittle to ductile failure. Above the transition depth, fracture occurs where the applied stresses exceed the rock's internal strength, generating earthquakes, while below this depth rocks creep ductilely.

Because both temperature and pressure increase with depth, the BDT marks the point where the weakening of rock's strength due to higher temperatures balances the strengthening of rocks due to higher pressures as illustrated in the figure below:
