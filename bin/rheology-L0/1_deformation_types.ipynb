## Rheology

Rheology is the study of the *flow and deformation in rocks* due to an applied stress. Different rocks deform differently based on their chemical structure and the physical state (temperature, pressure, strain rate, etc.) under which they are deformed. The relationship between the applied stress and the resultant deformation (strain, strain rate) is governed by the *constitutive laws*.

### Types of deformation

The three major types of deformation mechanisms are:

1) **Elastic**:
- A material deforms elastically if it goes back to its original shape once the applied force is removed.
- Deformation occurs instantaneously, and is reversible
- Described by Hooke's law as $\sigma = E\varepsilon$, where $\sigma$ and $\varepsilon$ are the applied stress and the generated strain, respectively, in the rock, and $E$ is the rock's Youngs' Modulus
- Is most common in low-pressure, low-stress conditions, i.e., near the Earth's surface
- It is most relevant for seismic wave propagation during an earthquake

2) **Plastic**:
- Stress-limited deformation; rocks fail once applied stresses exceed a yield criterion.
- Deformation is localized and occurs over short (brittle) or intermediate (ductile) timescales, and is permanent.
- Described by Drucker-Prager, Mohr-Coulomb failure criterion
- The accumulated strain increases with time while the material is yielding.
- It is most commonly manifested in high shear zones near the Earth's surface such as plate boundaries, faults, folds, etc.  

3) **Viscous**:
- At high temperatures and pressures, a rock can behave as a fluid and flow viscously.
- Deformation is permanent and time-dependent.
- Deformation rate of the material is linear (Newtonian) or nonlinear (non-Newtonian) relative to the applied stress.
- For a Newtonian fluid, strain rate ($\dot\varepsilon$) varies linearly in response to the applied stress as $\sigma = 2\eta\dot\varepsilon$, where $\eta$ is the fluid viscosity.
- Relevant for long term deformation such as mantle convection, seismic anisotropy, post-seismic relaxation, etc.
