### Clausius-Clapeyron relation

On a pressure-temperature (P-T) diagram for any phase change, the curve separating the two phases can be thought as the P-T space where both phases coexist. The Clapeyron slope, $dP/dT$, gives the slope of the tangent to the curve, and can be given mathematically as:

<div align="center">

$\dfrac{dP}{dT} = \dfrac{L}{T\Delta v} = \dfrac{\Delta s}{\Delta v},$

</div>

where $L$ is the latent heat, i.e., amount of heat absorbed (negative) or released (positive) during the phase change, and $\Delta s$,  $\Delta v$ is the molar entropy and volume change from the phase transition, respectively.

#### Clapeyron slope in upper mantle discontinuities
From the equation above, a positive Clapeyron slope indicates an exothermic phase change reaction and a negative slope and endothermic reaction. The observed 410 km seismic discontinuity has Clapeyron slope values of $\sim 2 \text{MPa K}^{-1}$.  Conversely, the 660 km seismic discontinuity has a negative Clapeyron slope of $\sim -2.5 \text{MPa K}^{-1}$.

### Phase changes in a subducting slab

Since a subducting slab is colder than the mantle around it, phase changes within the slab occurs at different depths than they do outside of it depending on the Clapeyron slope values:

* **410**: The slope is positive. In the cold slab, this transition happens at lower pressure/shallower depth. At these depths, the slab material is made up of a denser phase than the surrounding material, promoting the further sinking of the slab through the 410 km.

* **660**: The slope is negative. In the cold slab, this transition happens at deeper depth. Therefore at 660 km,the surrounding mantle has already transformed into the denser phase, creating an upward buoyancy that resists its descent through 660 km.

### Phase changes in a rising plume

Since a plume is hotter than the mantle around it, phase changes within the plume occurs at different depths than they do outside of it depending on the Clapeyron slope values:

* **410**: The slope is positive. In the hot plume, this transition happens at higher pressure/deeper depths. At these depths, the plume material is still in its shallower, less-dense phase while the surrounding mantle has already transitioned to a denser phase. This makes the plume even more buoyant, promoting the further rise of the plume through the 410 km boundary.

* **660**: The slope is negative. In the hot plume, this transition happens at a shallower depth (lower pressure). This means that just above 660 km, the plume has already transformed into the denser, deeper phase (perovskite), while the surrounding mantle is still in the lighter phase. Because the plume is now locally denser than its surroundings, it is negatively buoyancy that impedes the further rise of the plume through the 660 km boundary.
